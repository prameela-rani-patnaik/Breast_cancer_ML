```python
df.replace({'diagnosis': {'B': 0, 'M':1}},inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Drop the 'id' column as it's just an identifier, keeping features and 'diagnosis'
df_features = df.drop(columns=["id"])

# 2. Calculate the correlation matrix
corr_matrix = df_features.corr()

# 3. Set up the matplotlib figure size
plt.figure(figsize=(20, 16))

# 4. Draw the heatmap
sns.heatmap(
    corr_matrix,
    annot=False,          # Set to True if you want to see the raw numbers inside squares
    cmap="coolwarm",      # Blue = negative correlation, Red = positive correlation
    fmt=".2f",            # Formats annotation numbers to 2 decimal places
    linewidths=0.5,       # Puts a tiny gap between cells for readability
    vmax=1.0, vmin=-1.0   # Standardizes the color bar scale from -1 to 1
)

# 5. Add titles and show
plt.title("Feature Correlation Heatmap (Breast Cancer Dataset)", fontsize=18, pad=20)
plt.show()
```

```python
# target
X = df.drop(columns=["id",'diagnosis'],axis=1)
Y = df['diagnosis']
```

```python
#train-test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

```python
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_scaled=sc.fit_transform(X_train)
X_test_scaled=sc.transform(X_test)
```

```python
#training the models
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
#lr
lr=LogisticRegression()
lr.fit(X_train_scaled,Y_train)
lr_pred=lr.predict(X_test_scaled)
print(lr_pred)
lr_acc=accuracy_score(Y_test,lr_pred)
print(lr_acc)
#knn
knn=KNeighborsClassifier()
knn.fit(X_train_scaled,Y_train)
knn_pred=knn.predict(X_test_scaled)
print(knn_pred)
knn_acc=accuracy_score(Y_test,knn_pred)
print(knn_acc)
#svc
svc=SVC(probability=True)
svc.fit(X_train_scaled,Y_train)
svc_pred=svc.predict(X_test_scaled)
print(svc_pred)
svc_acc=accuracy_score(Y_test,svc_pred)
print(svc_acc)
#rf
rf=RandomForestClassifier()
rf.fit(X_train_scaled,Y_train)
rf_pred=rf.predict(X_test_scaled)
print(rf_pred)
rf_acc=accuracy_score(Y_test,rf_pred)
print(rf_acc)
#dt
dt=DecisionTreeClassifier()
dt.fit(X_train_scaled,Y_train)
dt_pred=dt.predict(X_test_scaled)
print(dt_pred)
dt_acc=accuracy_score(Y_test,dt_pred)
print(dt_acc)
```

```python
#classification report
from sklearn.metrics import classification_report
print(classification_report(Y_test,lr_pred))
print(classification_report(Y_test,knn_pred))
print(classification_report(Y_test,svc_pred))
print(classification_report(Y_test,rf_pred))
print(classification_report(Y_test,dt_pred))
#
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

models = {
    "Logistic Regression": lr_pred,
    "KNN": knn_pred,
    "SVM": svc_pred,
    "Random Forest": rf_pred,
    "Decision Tree": dt_pred
}

for name, pred in models.items():
    cm = confusion_matrix(Y_test, pred)

    ConfusionMatrixDisplay(cm).plot(cmap='Blues')
    plt.title(name)
    plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Dictionary containing the positive class probabilities for each model
model_probs = {
    "Logistic Regression": lr_pred_prob[:, 1],
    "KNN": knn_pred_prob[:, 1],
    "SVM": svc_pred_prob[:, 1],
    "Random Forest": rf_pred_prob[:, 1],
    "Decision Tree": dt_pred_prob[:, 1]
}

# Set up the plot
plt.figure(figsize=(10, 4))

# Loop through each model to calculate and plot its ROC curve
for name, probs in model_probs.items():
    fpr, tpr, _ = roc_curve(Y_test, probs)
    # Changed label here to only display the model name
    plt.plot(fpr, tpr, label=name, lw=2)

# Plot the diagonal baseline (random guessing)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Guess')

# Formatting the graph
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1 - Specificity)', fontsize=12)
plt.ylabel('True Positive Rate (Sensitivity / Recall)', fontsize=12)
plt.title('ROC Curve Comparison (All Models)', fontsize=14)
plt.legend(loc="lower right", fontsize=11)
plt.grid(True, linestyle=':', alpha=0.6)

# Display the plot
plt.show()
```

```python
#roc-auc score
from sklearn.metrics import roc_auc_score
lr_roc_auc_score=roc_auc_score(Y_test,lr_pred)
knn_roc_auc_score=roc_auc_score(Y_test,knn_pred)
svc_roc_auc_score=roc_auc_score(Y_test,svc_pred)
rf_roc_auc_score=roc_auc_score(Y_test,rf_pred)
dt_roc_auc_score=roc_auc_score(Y_test,dt_pred)
print("lr:",lr_roc_auc_score)
print("knn:",knn_roc_auc_score)
print("svm:",svc_roc_auc_score)
print("rf:",rf_roc_auc_score)
print("dt:",dt_roc_auc_score)
```

```python
import pandas as pd

# Recall scores for class 1 (Malignant) as manually extracted from classification reports
# These values are crucial for medical diagnosis as missing a positive case is high risk.
recall_malignant = {
    "Logistic Regression": 0.98,
    "KNN": 0.93,
    "SVM": 0.98,
    "Random Forest": 0.93,
    "Decision Tree": 0.98  # Updated based on latest classification report for DT
}

# Create a dictionary to store all metrics
summary_data = {
    'Model': ['Logistic Regression', 'KNN', 'SVM', 'Random Forest', 'Decision Tree'],
    'Accuracy': [lr_acc, knn_acc, svc_acc, rf_acc, dt_acc],
    'Recall (Malignant)': [recall_malignant['Logistic Regression'],
                           recall_malignant['KNN'],
                           recall_malignant['SVM'],
                           recall_malignant['Random Forest'],
                           recall_malignant['Decision Tree']],
    'ROC-AUC Score': [lr_roc_auc_score, knn_roc_auc_score, svc_roc_auc_score, rf_roc_auc_score, dt_roc_auc_score]
}

# Create a DataFrame from the summary data
summary_df = pd.DataFrame(summary_data)

# Display the summary table
display(summary_df.round(3))
```

```python
model_preference = summary_df.sort_values(by=['Recall (Malignant)', 'ROC-AUC Score', 'Accuracy'], ascending=[False, False, False])

print("\nPreferred Order of Models (ranked by Recall (Malignant), then ROC-AUC, then Accuracy):\n")
for index, row in model_preference.iterrows():
    print(f"{index + 1}. {row['Model']}")
```
